```python
import os
import requests
import folium
from getpass import getpass
from langchain.agents import initialize_agent, Tool, AgentType
from langchain.chat_models import ChatOpenAI


# Set up environment variables for API keys
def _set_env(var: str):
    if not os.getenv(var):
        os.environ[var] = getpass(f"Enter {var}: ")


_set_env("OPENAI_API_KEY")
GOOGLE_API_KEY = getpass("Enter your Google Maps API key: ")
YOUTUBE_API_KEY = getpass("Enter your YouTube Data API key: ")

# Define tools
def get_nearby_facilities(api_key, location, radius=5000, interest=""):
    geocode_url = f"https://maps.googleapis.com/maps/api/geocode/json?address={location}&key={api_key}"
    geocode_response = requests.get(geocode_url).json()

    if geocode_response.get("results"):
        coordinates = geocode_response["results"][0]["geometry"]["location"]
        latitude = coordinates["lat"]
        longitude = coordinates["lng"]

        places_url = (
            f"https://maps.googleapis.com/maps/api/place/nearbysearch/json?"
            f"location={latitude},{longitude}&radius={radius}&keyword={interest}&key={api_key}"
        )
        places_response = requests.get(places_url).json()

        if places_response.get("results"):
            return places_response["results"], (latitude, longitude)
        else:
            return [], (latitude, longitude)
    else:
        return [], None


def get_youtube_practice_videos(youtube_api_key, interest, max_results=5):
    search_url = (
        f"https://www.googleapis.com/youtube/v3/search?"
        f"part=snippet&q={interest} practice drills&type=video&maxResults={max_results}&key={youtube_api_key}"
    )
    response = requests.get(search_url).json()

    if "items" in response:
        return response["items"]
    else:
        return []


# Define the tools
tools = [
    Tool(
        name="Find Nearby Facilities",
        func=lambda query: get_nearby_facilities(GOOGLE_API_KEY, query, interest="cricket"),
        description="Find nearby facilities based on user input location.",
    ),
    Tool(
        name="Get YouTube Practice Videos",
        func=lambda query: get_youtube_practice_videos(YOUTUBE_API_KEY, query),
        description="Get YouTube practice videos for a given interest.",
    ),
]

# Initialize the LLM
llm = ChatOpenAI(
    model="gpt-4-turbo",
    temperature=0,
    openai_api_key=os.environ["OPENAI_API_KEY"],
)

# Initialize the agent
agent = initialize_agent(
    tools=tools,
    llm=llm,
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True,
)

# Function to display facilities on a map
def display_map(grounds, user_location):
    map_obj = folium.Map(location=user_location, zoom_start=13)
    folium.Marker(user_location, popup="Your Location", icon=folium.Icon(color="blue")).add_to(map_obj)

    for ground in grounds:
        name = ground.get("name")
        lat = ground["geometry"]["location"]["lat"]
        lng = ground["geometry"]["location"]["lng"]
        address = ground.get("vicinity", "Address not available")
        folium.Marker(
            [lat, lng],
            popup=f"{name}\n{address}",
            icon=folium.Icon(color="green"),
        ).add_to(map_obj)

    map_obj.save("nearby_facilities.html")
    print("Map saved as 'nearby_facilities.html'. Open it in your browser to view.")


# Main function
if __name__ == "__main__":
    user_location = input("Enter your location (e.g., city or latitude,longitude): ")
    user_interest = input("Enter your interest (e.g., cricket, football, tennis): ")

    # Agent handles planning and execution
    response = agent.run(f"Find {user_interest} facilities near {user_location} and suggest practice videos.")
    print("\nAgent Response:\n", response)

    # Retrieve facilities and videos separately
    results, user_coordinates = get_nearby_facilities(GOOGLE_API_KEY, user_location, interest=user_interest)
    if results:
        display_map(results, user_coordinates)
        videos = get_youtube_practice_videos(YOUTUBE_API_KEY, user_interest)
        print("\nYouTube Practice Drills:")
        for video in videos:
            title = video["snippet"]["title"]
            video_url = f"https://www.youtube.com/watch?v={video['id']['videoId']}"
            print(f"- {title}: {video_url}")
```

```python
import feedparser

def get_sports_news(user_interest):
    # Define the RSS feed URL
    rss_url = 'https://feeds.bbci.co.uk/sport/rss.xml'

    # Parse the RSS feed
    feed = feedparser.parse(rss_url)

    # List to store articles matching user interest
    matching_articles = []

    # Iterate over each entry (article) in the feed
    for entry in feed.entries:
        # Check if the user interest is in the article title or description
        if user_interest.lower() in entry.title.lower() or user_interest.lower() in entry.description.lower():
            article = {
                'title': entry.title,
                'link': entry.link,
                'description': entry.description
            }
            matching_articles.append(article)

    return matching_articles

# Example usage
user_interest = input("Enter your sport interest (e.g., 'football', 'tennis'): ")
articles = get_sports_news(user_interest)

if articles:
    print(f"Found {len(articles)} articles related to '{user_interest}':\n")
    for article in articles:
        print(f"Title: {article['title']}")
        print(f"Link: {article['link']}")
        print(f"Description: {article['description']}\n")
else:
    print(f"No articles found related to '{user_interest}'.")
```
